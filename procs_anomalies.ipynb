# Procs anomaly detection
Try and find anomalies within the procs of a single job.
Look for: repeating numbers. drops to zero. repeated actions. jumps in location.

```python
#import packages
import epmt_query as eq
import sys
sys.path.insert(0,'/home/Avery.Kiihne/pip_experiment')
import matplotlib.pyplot as plt
import numpy as np
import time
import matplotlib.animation as animation
```

```python
#use orm format to get jobs and then feed to get_procs for a quicker run time
#need few jobs because each will be lasered in on
start = time.time()
job_num = 1
jobs_all = eq.get_jobs(limit = job_num,
                       before = -8,
                       fmt='orm') #format for passing to get_procs
procs_all = eq.get_procs(jobs = jobs_all,
                       fmt='dict')
print(f'number of elements in procs_all={len(procs_all)}')
print("time:",(time.time()-start)/60)
```

```python
#create one large forloop for whole job, that conducts a number of tests in one loop
read_bytes_record = []
write_bytes_record = []
wait_time_record = []
rssmax_record = []

for proc in procs_all:
    if proc['read_bytes'] > 0:
        read_bytes_record.append(proc['read_bytes'])
    if proc['write_bytes'] >0:
        write_bytes_record.append(proc['write_bytes'])
    if proc['time_waiting'] >0:
        wait_time_record.append(proc['time_waiting'])
    if proc['rssmax'] >0:
        rssmax_record.append(proc['rssmax'])
```
