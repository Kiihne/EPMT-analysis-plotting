```python
#this notebook contains plotting commands for reference
import numpy as np
import time
#Have to use special path to find plt in my system
import sys
sys.path.insert(0,'/home/Avery.Kiihne/pip_experiment')
import matplotlib.pyplot as plt
```

```python
start = time.time() #this will give a runtime for the code
print('starting')

x = np.linspace(0,10,21) #make variable range (start,end,number of digits)
y = x**2-1
#plotting
plt.plot(x,y, color = 'b', label = 'log')
plt.plot(x,x, color = 'red',alpha = .6, label = 'linear') #alpha is transparancy

#axis
plt.yscale('log') #log or linear
plt.ylim(.1,110) #ylim cannot = 0 if log scale
plt.xlim(-1,11)
plt.tick_params(direction='out', length=10, labelsize = 15, color = 'green', which = 'major')
plt.tick_params(direction='out', length=5, color = 'orange', which = 'minor')
plt.yticks((1e-1,1e0,1e1,1e2)) #set exact axis markers

#markups
plt.style.use('default')  #this exists to counter any local chnages made to coloring, such as dark mode
plt.title('Example One',color='black', fontsize=15, weight ='bold',horizontalalignment='center',verticalalignment='bottom', bbox=dict(facecolor='white', alpha=0.8))
plt.annotate('annotation',(-.5,15)) #put words or boxes on plot. (x,y) cordinates
plt.legend() #use label when legend should look like lines printed

plt.tight_layout()   #magical command that can fix alot of random spacing issues

#plt.savefig('example_plot', bbox_inches='tight', format = 'pdf')

end = time.time()
print("time:",(end-start)/60)
```
